```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
# source = requests.get('https://www.iitdh.ac.in/bsbe.php').text
source = requests.get('https://en.wikipedia.org/wiki/IIT_Dharwad').text
soup = BeautifulSoup(source, 'lxml')
```

```python
print(soup.prettify())
print(soup.get_text())
```

```python
p1 = soup.find('p')
print(p1.get_text())
```

```python
p2 = soup.find_all('p', class_='font_medum text-justify')
print(p2)
```

```python
p3 = "".join([p.text for p in soup.find_all("p", class_='font_medum text-justify')])
print(p3)
```

```python
# links = "".join([p.text for p in soup.find_all("a")])
link = soup.find('a')
print(link.get('href'))
```
