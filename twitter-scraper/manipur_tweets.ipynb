```python
from twikit import Client
import csv
```

```python

USERNAME = 'DataC51466'
EMAIL = 'xxxxxxxxxxxxxxx'
PASSWORD = 'xxxxxxx'
# Initialize client
client = Client('en-US')
# Login to the service with provided user credentials
client.login(
    auth_info_1=USERNAME ,
    auth_info_2=EMAIL,
    password=PASSWORD
)
```

```python
tweets = client.search_tweet('Manipur', 'Latest')
for tweet_id in tweets:
    print(tweet_id)
# Search more tweets
more_tweets = tweets.next()
for tweet_id in more_tweets:
    print(tweet_id)
```

```python
with open('manipur_tweets.csv', 'w', newline='') as output_csv_file:
    csv_writer = csv.writer(output_csv_file)
    csv_writer.writerow(['id', 'text', 'favorite_count'])    
    for tweet in tweets:
        tweet = client.get_tweet_by_id(tweet.id)
        # Access tweet attributes
        print(
            f'id: {tweet.id}',
            f'text {tweet.text}',
            f'favorite count: {tweet.favorite_count}',
            f'media: {tweet.media}',
            sep='\n'
        )
        csv_writer.writerow([tweet.id, tweet.text, tweet.favorite_count])
```
