```python
import pytesseract
from pdf2image import convert_from_path
```

```python

pdf_path = 'sample_pdfs/sample1.pdf'

images = convert_from_path(pdf_path)
extracted_text = ''
for img in images:
    text = pytesseract.image_to_string(img, lang='ben')
    extracted_text += text
print(extracted_text)
```
