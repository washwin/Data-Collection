```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = "https://epaper.thesangaiexpress.com/index.php?edition=Mnppage&date=2024-03-04&page=1"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify)
```

```python
image_tags = soup.find_all('img')
png_urls = []
for img in image_tags:
    src = img.get('src')
    if src and ".png" in src:
        png_urls.append(src)

for png_url in png_urls:
    print(png_url)
```
